Abaixo segue um gráfico demonstrando o experimento dado a repetição atual e o número de casos de ganho ao trocar e ao não trocar de porta:

```python
import matplotlib.pyplot as plt
import simulacao as sim
import numpy as np

plt.figure(figsize=(16,9))

xNaoTroca = sim.getPlotDataNaoTroca()[0]
yNaoTroca = sim.getPlotDataNaoTroca()[1]
xTroca = sim.getPlotDataTroca()[0]
yTroca = sim.getPlotDataTroca()[1]

trocou = plt.plot(xTroca, yTroca, label='Trocou de Porta')
naoTrocou = plt.plot(xNaoTroca,yNaoTroca, label='Não Trocou de Porta')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axhline(y = 1/3, linewidth = 1, color='grey', linestyle = '--')
plt.axhline(y = 2/3, linewidth = 1, color='grey', linestyle = '--')

plt.ylim(0, 1)    
plt.xlim(0, sim.getTotal())
plt.xlabel("Quantidade de Simulações")
plt.ylabel("Probabilidade de Premiação")
plt.xticks(np.arange(0, sim.getTotal() + 1, 500))
plt.yticks(np.arange(0, 1.0001, 0.05))

plt.show()
```

Temos que as linhas tracejadas em cor cinza representam 1/3 do limite de repetições (10000), portanto, podemos concluir graficamente que a chance de vitória ao trocar de porta é de 2/3, enquanto a chance do participante vencer não trocando de porta é de 1/3. (Valores teóricos).
